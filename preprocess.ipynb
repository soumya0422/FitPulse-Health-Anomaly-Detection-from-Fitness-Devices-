```python
import pandas as pd
import numpy as np
```

```python
hr = pd.read_csv('/content/drive/MyDrive/Data/heartrate_seconds_merged.csv')
steps = pd.read_csv('/content/drive/MyDrive/Data/hourlySteps_merged.csv')
sleep = pd.read_csv('/content/drive/MyDrive/Data/sleepDay_merged.csv')
```

```python
sleep['SleepDay'] = pd.to_datetime(sleep['SleepDay'],format='%m/%d/%Y %I:%M:%S %p')
sleep['Sleep_per_minute'] = sleep['TotalMinutesAsleep'] / (24 * 60)
sleep
```

```python
steps['ActivityHour'] = pd.to_datetime(steps['ActivityHour'],format='%m/%d/%Y %I:%M:%S %p')
steps['Steps_per_minute'] = steps['StepTotal'] / 60
steps['Date'] = steps['ActivityHour'].dt.date
steps
```

```python
hr['Time'] = pd.to_datetime(hr['Time'])
hr['Time'] = hr['Time'].dt.floor('min')
hr_1min = (
    hr
    .groupby(['Id', 'Time'], as_index=False)
    .agg(heart_rate=('Value', 'mean'))
)
hr_1min
```

```python
hr_1min['Date'] = hr_1min['Time'].dt.date
hr_1min['Hour'] = hr_1min['Time'].dt.hour
steps['Date'] = pd.to_datetime(steps['ActivityHour']).dt.date
steps['Hour'] = pd.to_datetime(steps['ActivityHour']).dt.hour
final_df = pd.merge(
    hr_1min,
    steps[['Id', 'Date', 'Hour', 'Steps_per_minute']],
    on=['Id', 'Date', 'Hour'],
    how='left'
)
final_df
```

```python
final_df['Date'] = pd.to_datetime(final_df['Date']).dt.date
sleep['Date'] = pd.to_datetime(sleep['SleepDay']).dt.date
```

```python

sleep['Date'] = pd.to_datetime(sleep['SleepDay']).dt.floor('D')
sleep
```

```python
final_df['Date'] = pd.to_datetime(final_df['Date'])
final_merged = pd.merge(
    final_df,
    sleep[['Id', 'Date', 'Sleep_per_minute']],
    on=['Id', 'Date'],
    how='left'
)
final_merged
```
